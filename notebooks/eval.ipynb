## Set prompt

```python
SYSTEM_PROMPT = """
You are an expert judge for evaluating answers to questions.
You will evaluate the answers to interview questions in Java domain.

Score the candidate's answer from 1 to 5.
5 = Completely correct, precise, and clear.
4 = Mostly correct, minor omission.
3 = Partially correct, but missing important details.
2 = Mostly incorrect, small parts may be true.
1 = Completely incorrect or irrelevant.

Your answer should be only an integer score from 1 to 5.
"""
```

```python
PROMPT = """
Here is the question, and the answer to it from the user:
Question: What are the four main principles of Object-Oriented Programming?
Answer: They are: abstraction, encapsulation, inheritance, polymorphism
"""
```

```python
messages = [
    {"role": "system", "content": SYSTEM_PROMPT},
    {"role": "user", "content": PROMPT}
]
```

```python
response = llm.query_ai(messages)
```
